```python
%sql
USE CATALOG football_catalog;
CREATE SCHEMA IF NOT EXISTS gold;

-- 1. Premier League: Top 5 Teams (Dynamic)
CREATE OR REPLACE TABLE gold.top_pl_season AS
SELECT team1 AS team, COUNT(*) as home_wins
FROM silver.pl_matches
WHERE s1 > s2
GROUP BY team1
ORDER BY home_wins DESC
LIMIT 5;

-- 2. La Liga: Top 5 Teams (Dynamic)
CREATE OR REPLACE TABLE gold.top_laliga_season AS
SELECT team1 AS team, COUNT(*) as home_wins
FROM silver.laliga_matches
WHERE s1 > s2
GROUP BY team1
ORDER BY home_wins DESC
LIMIT 5;
```

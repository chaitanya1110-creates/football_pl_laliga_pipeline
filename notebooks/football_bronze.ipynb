```python
import requests
import json

# 1. SETUP: Catalog and Schema
spark.sql("CREATE CATALOG IF NOT EXISTS football_catalog")
spark.sql("USE CATALOG football_catalog")
spark.sql("CREATE SCHEMA IF NOT EXISTS bronze")

# 2. SOURCES: Premier League and La Liga (GitHub Raw)
urls = {
    "pl_2026": "https://raw.githubusercontent.com/openfootball/football.json/master/2025-26/en.1.json",
    "laliga_2026": "https://raw.githubusercontent.com/openfootball/football.json/master/2025-26/es.1.json"
}

# 3. FETCH & SAVE: Simplest possible ingestion
for name, url in urls.items():
    raw_payload = requests.get(url).text
    df = spark.createDataFrame([(raw_payload,)], ["payload"])
    df.write.mode("overwrite").saveAsTable(f"bronze.raw_{name}")

print("Bronze: All leagues ingested.")
```
